```python
import random
import math as m
import statistics as st
import numpy as np


class Mobile_soul:
    
    def __init__(self):
        self.size = 5
        self.max_speed = 33
        self.current_speed = 1
        self.position = 0
        
    def motion(self):
        if self.current_speed < 0:
            self.position += 0
        else:
            self.position += self.current_speed
    
    def accelerate(self):
        if self.current_speed < self.max_speed:
            self.current_speed += 2
        else:
            self.current_speed += 0

    def decelerate(self):
        chance = random.randint(1,11)
        if chance == 1:
            self.current_speed -= 2
    
    #def flow_check(self):
        #if self.current_speed == self.max_speed:
            
    
    
class Soul_river:
    
    def __init__(self, step):
        self.river_length = self.splash(step)
        self.soul = [Mobile_soul() for _ in self.river_length]
        self.soul_2 = self.posit()
       
        
    def posit(self):
        t_list = []
        zipper_2 = [list(i) for i in list(zip(self.soul, self.river_length))]
        for i in zipper_2:
            i[0].position += i[1]
            t_list.append(i[0])
        return t_list
            
    def splash(self, step): # river divide
        n_list = []
        n_step = range(step)
        o_list = [_ * (1000/step) for _ in range(0, step)]
        for i in o_list:
            n_list.append(int(i))
        return n_list
    
```

```python
soul = Mobile_soul()
riv = Soul_river(5)
print(riv.soul_2)

def entelocheia():
    for i in riv.soul_2:
        print("speed: {}".format(i.current_speed))
        i.motion()
        i.accelerate()
        i.decelerate()
        print(i.position)
        if i.position > 1000:
            i.position -= 1000
            
for i in range(60):
    entelocheia()
```
